1, Reduce the data dimensionality to 2 dimensions using PCA and display the data after dimensionality reduction, visualize the data and also normalize the data

```python
# Chia tập dữ liệu thành các biến phụ thuộc và độc lập

x = wine.iloc[:, :11]
y = wine.iloc[:, 11]

print(x.shape)
print(y.shape)
```

```python
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=7)

# determining the shapes of training and testing sets
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
```

```python
# Chuẩn hóa dữ liệu 
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

from sklearn.ensemble import ExtraTreesClassifier

classifiern = ExtraTreesClassifier()
classifiern.fit(x, y)
score = classifiern.feature_importances_
print(score)
```

```python
from sklearn.decomposition import PCA

# Khởi tạo và huấn luyện mô hình PCA
pca = PCA(n_components=2)  # Chỉ định số chiều mới là 2
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
# Hiển thị dữ liệu sau khi giảm chiều trên biểu đồ phân tán
plt.figure(figsize=(8, 6))
plt.scatter(x_train_pca[:, 0], x_train_pca[:, 1], c=y_train, cmap='viridis')
plt.title('PCA Reduced Data Scatter Plot')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target')
plt.show()
```

2, Apply linear regression, with rounding the model output, train the model with training data and then try to predict on test data. Check the model's suitability through the accuracy (loss) index.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, mean_squared_error

# Huấn luyện mô hình hồi quy tuyến 
model = LinearRegression()
model.fit(x_train, y_train)

# Dự đoán đầu ra cho dữ liệu test
Y_pred = model.predict(x_test)

# Làm tròn đầu ra của dự đoán
Y_pred_rounded = [round(y) for y in Y_pred]

# Đánh giá mô hình 
mse = mean_squared_error(y_test, Y_pred_rounded)

print("Mean Squared Error: ", mse)

# Tính toán độ chính xác (accuracy) của mô hình
accuracy = accuracy_score(y_test, Y_pred_rounded)
print("Accuracy:", accuracy)

# Vẽ biểu đồ so sánh giá trị thực và giá trị dự đoán
plt.figure(figsize=(10, 6))
plt.scatter(range(len(y_test)), y_test, color='blue', label='Actual')
plt.scatter(range(len(y_test)), Y_pred_rounded, color='red', label='Predicted')
plt.xlabel('Sample Index')
plt.ylabel('Quality')
plt.title('Actual vs Predicted Quality (Rounded Predictions)')
plt.legend()
plt.show()
```

3, Put the original problem into data classification form. With the following ideas:
    a, Use classification models: logistic regression, Naïve Bayes matching; to perform this classification problem. Compare
    methods on the following criteria: Accuracy, precision, recall and running time.

Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
import time

# Khởi tạo và huấn luyện mô hình Logistic Regression
model = LogisticRegression()
start_time = time.time()
model.fit(x_train, y_train)
end_time = time.time()
runtime = end_time - start_time

# Dự đoán biến mục tiêu trên tập dữ liệu kiểm tra
Y_pred1 = model.predict(x_test)

# Tính các độ đo precision và recall
precision = precision_score(y_test, Y_pred1)
recall = recall_score(y_test, Y_pred1)

print("Accuracy Score:", accuracy_score(y_test, Y_pred1))
print("Precision Score:", precision)
print("Recall Score:", recall)
print("Runtime:", runtime, "seconds")

print(classification_report(y_test, Y_pred1))
confusion_mat = confusion_matrix(y_test, Y_pred1)
print(confusion_mat)
```

Using Navie Bayes

```python
from sklearn.naive_bayes import GaussianNB

# Khởi tạo và huấn luyện mô hình Navie Bayes (dung gaussian)
model = GaussianNB()
start_time = time.time()
model.fit(x_train, y_train)
end_time = time.time()
runtime = end_time - start_time

# Dự đoán biến mục tiêu trên tập dữ liệu kiểm tra
y_pred = model.predict(x_test)

# Tính các độ đo precision và recall
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("Precision Score:", precision)
print("Recall Score:", recall)
print("Runtime:", runtime, "seconds")

print(classification_report(y_test, y_pred))
confusion_mat = confusion_matrix(y_test, y_pred)
print(confusion_mat)
```

    b, Reduce the number of dimensions to half (round up) the current number using PCA. Then again use two stool models
    the above mentioned type to resolve.Compare the original data case and the reduced data case. Note the need to reduce dimension
    on synthetic data including training and testing.

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=7)
# Giảm số chiều xuống còn một nửa (làm tròn lên) số chiều hiện tại bằng PCA
original_dim = x.shape[1]
reduced_dim = int(np.ceil(original_dim / 2))

pca = PCA(n_components=reduced_dim)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Vẽ biểu đồ phân tán cho dữ liệu PCA
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=Y_train, cmap='viridis', alpha=0.5, edgecolor='k',
            label='Training Data')
plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=Y_test, cmap='viridis', alpha=0.8, edgecolor='k', marker='*',
            label='Test Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Reduced Data Scatter Plot')
plt.colorbar(label='Target')
plt.legend()
plt.show()
```

Using Logistic Regression after reduce dimensions

```python
model = LogisticRegression()
start_time = time.time()
model.fit(X_train, Y_train)
end_time = time.time()
runtime = end_time - start_time

# Dự đoán biến mục tiêu trên tập dữ liệu kiểm tra
Y_pred = model.predict(X_test)

# Tính các độ đo precision và recall
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)

print("Accuracy Score:", accuracy_score(Y_test, Y_pred))
print("Precision Score:", precision)
print("Recall Score:", recall)
print("Runtime:", runtime, "seconds")

print(classification_report(Y_test, Y_pred))
confusion_mat = confusion_matrix(Y_test, Y_pred)
print(confusion_mat)
```

Using Navie Bayes after reduce dimensions

```python
model = GaussianNB()
start_time = time.time()
model.fit(X_train, Y_train)
end_time = time.time()
runtime = end_time - start_time

# Dự đoán biến mục tiêu trên tập dữ liệu kiểm tra
y_pred = model.predict(X_test)

# Tính các độ đo precision và recall
precision = precision_score(Y_test, y_pred)
recall = recall_score(Y_test, y_pred)

print("Accuracy Score:", accuracy_score(Y_test, y_pred))
print("Precision Score:", precision)
print("Recall Score:", recall)
print("Runtime:", runtime, "seconds")

print(classification_report(Y_test, y_pred))
confusion_mat = confusion_matrix(Y_test, y_pred)
print(confusion_mat)
```
